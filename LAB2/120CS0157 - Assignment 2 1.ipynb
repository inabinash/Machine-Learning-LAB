```python
# necessary imports
import math
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split 
```

####  1. Generate N = 500, 2-D random data points and plot its corresponding Gaussian PDF.

```python
def generateGaussianPDF(mean, cov, size):
    
    # mean is a 2-d matrix
    # cov is 2-d symmmetric matrix
    # size is data size
    
    x, y = np.random.multivariate_normal(mean, cov, size).T
    return x, y

def plotGaussianPDF(x, y):
    plt.plot(x, y, 'g.')
    
    # setting range of x and y axis
    plt.axis([-10, 10, -10, 10])
    
    plt.show()
    
x, y = generateGaussianPDF([0, 0], [[1, 0], [0, 1]], 500)
plotGaussianPDF(x, y)
```

#### 2. Generate N = 500, 2-dimensional data points that are distributed according to the Gaussian distributionN(m,S), with mean m = [0, 0]T and covariance matrix 
        S1=σ2 1 σ12 σ12 σ2 2


```python
# given
mean = [0, 0]
cov = [
    [
        [1, 0],
        [0, 1]
    ],
    [
        [0.2, 0],
        [0, 0.2]
    ],
    [
        [2, 0],
        [0, 2]
    ],
    [
        [0.2, 0],
        [0, 2]
    ],
    [
        [2, 0],
        [0, 0.2]
    ],
    [
        [1, 0.5],
        [0.5, 1]
    ],
    [
        [0.3, 0.5],
        [0.5, 2]
    ],
    [
        [0.3, -0.5],
        [-0.5, 2]
    ]
]
```

```python
for i in range(len(cov)):
    x, y = generateGaussianPDF(mean, cov[i], 500)
    print("Plot : ", i+1)
    plotGaussianPDF(x, y)
```

#### 3. Considering iris dataset - ”https://archive.ics.uci.edu/ml/datasets/Iris”, D of size N Ö M, with N: number of samples and M: number of features, design a Bayesian classiﬁer to classify the test data. Divide the dataset into training and testing data into (a) 70:30 ratio (b) 80:20 ratio, calculating the accuracy and error rate

```python
data = pd.read_csv("iris.data")
```

```python
data
```

#### (a)

```python
x = data.iloc[:,:4]
y = data.iloc[:,-1]
X_train, X_test, Y_train, Y_test = train_test_split(x, y, test_size = 0.3)
model = GaussianNB()
model.fit(X_train, Y_train)
predictions = model.predict(X_test)
accuracy = accuracy_score(predictions, Y_test)
print("Accuracy: ", accuracy)
```

#### (b)

```python
x = data.iloc[:,:4]
y = data.iloc[:,-1]
X_train, X_test, Y_train, Y_test = train_test_split(x, y, test_size = 0.2)
model = GaussianNB()
model.fit(X_train, Y_train)
predictions = model.predict(X_test)
accuracy = accuracy_score(predictions, Y_test)
print("Accuracy: ", accuracy)
```
